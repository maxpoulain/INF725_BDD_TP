```python
import sqlite3
```

```python
def printSchema(connection):
    ### Source: http://stackoverflow.com/a/35092773/4765776
    for (tableName,) in connection.execute(
        """
        select NAME from SQLITE_MASTER where TYPE='table' order by NAME;
        """
    ):
        print("{}:".format(tableName))
        for (
            columnID, columnName, columnType,
            columnNotNull, columnDefault, columnPK,
        ) in connection.execute("pragma table_info('{}');".format(tableName)):
            print("  {id}: {name}({type}){null}{default}{pk}".format(
                id=columnID,
                name=columnName,
                type=columnType,
                null=" not null" if columnNotNull else "",
                default=" [{}]".format(columnDefault) if columnDefault else "",
                pk=" *{}".format(columnPK) if columnPK else "",
            ))
```

__1.3__ Create the new tables from Master1:

```python
# Remove previous results
%sql DROP TABLE IF EXISTS producteurs;
%sql DROP TABLE IF EXISTS vins;
%sql DROP TABLE IF EXISTS quantite;

# Create vins table
%sql CREATE TABLE vins AS \
SELECT DISTINCT NV, CRU, DEGRE, MILL \
FROM MASTER1 \
WHERE NV != 'None'\
ORDER BY NV ASC;

# Create producteurs table
%sql CREATE TABLE producteurs AS \
SELECT DISTINCT NP, NOM, PRENOM, REGION \
FROM MASTER1 \
WHERE NP != 'None'\
ORDER BY NP ASC;

# Create quantite table
%sql CREATE TABLE quantite AS \
SELECT DISTINCT NV, NP, QTE \
FROM MASTER1 \
WHERE NV != 'None'\
ORDER BY NV ASC;

print("\nContent of the database")
printSchema(conn)
```

__1.4__ Create the new tables from Master2:

```python
# Remove previous results
%sql DROP TABLE IF EXISTS vins;
%sql DROP TABLE IF EXISTS buveurs;
%sql DROP TABLE IF EXISTS achats;
%sql DROP TABLE IF EXISTS locations;

# Create vins table
%sql CREATE TABLE vins AS \
SELECT DISTINCT NV, CRU, DEGRE, MILL \
FROM MASTER2 \
WHERE NV != 'None'\
ORDER BY NV ASC;

# Create buveurs table
%sql CREATE TABLE buveurs AS \
SELECT DISTINCT NB, NOM, PRENOM, TYPE \
FROM MASTER2 \
WHERE NB != 'None'\
ORDER BY NB ASC;

# Create achats table
%sql CREATE TABLE achats AS \
SELECT DISTINCT NV, NB, QTE, DATES, LIEU \
FROM MASTER2 \
WHERE NV != 'None';

# Create locations table
%sql CREATE TABLE locations AS \
SELECT DISTINCT LIEU, REGION \
FROM MASTER2;

print("\nContent of the database")
printSchema(conn)
```

__2.2__ What regions produce Pommard or Brouilly?

```python
%%sql 
SELECT DISTINCT CRU,REGION 
FROM vins v, producteurs p, quantite q
WHERE p.NP = q.NP and v.NV = q.NV and CRU in ("Pommard","Brouilly") 
ORDER BY CRU;
```

__2.3__ What regions produce Pommard and Brouilly?

```python
%%sql 
SELECT DISTINCT REGION 
FROM vins v, producteurs p, quantite q
WHERE p.NP = q.NP and v.NV = q.NV and CRU="Pommard"
INTERSECT
SELECT DISTINCT REGION 
FROM vins v, producteurs p, quantite q
WHERE p.NP = q.NP and v.NV = q.NV and CRU="Brouilly"
```

__2.4__ Get the number of wines bught by CRU and Millésime

```python
%%sql 
SELECT CRU, MILL, SUM(QTE) AS Number_of_wines_bought 
FROM vins v, achats a 
WHERE v.NV = a.NV 
GROUP BY CRU, MILL
HAVING Number_of_wines_bought != 'None'
ORDER BY CRU, MILL desc;
```

__2.5__ Retrieve the wine number (NV) of wines produced by more than three producers

```python
%%sql 
SELECT * FROM (SELECT NV, COUNT(NP) AS Nb_of_producers FROM quantite q
GROUP BY NV)
WHERE Nb_of_producers > 3;
```

__2.6__ Which producers have not produced any wine?

```python
%%sql 
SELECT DISTINCT p.NP,NOM, PRENOM  FROM producteurs p
EXCEPT
SELECT DISTINCT p.NP,NOM, PRENOM FROM quantite q, producteurs p
WHERE q.NP = p.NP
ORDER BY p.NP;
```

__2.7__ What clients (buveurs) have bought at least one wine from 1980?

```python
%%sql 
SELECT DISTINCT b.NB, b.NOM, b.PRENOM FROM buveurs b, achats a, vins v
WHERE a.NB = b.NB and v.NV = a.NV and MILL = 1980
ORDER BY b.NB;
```

__2.8__ What clients (buveurs) have NOT bought any wine from 1980?

```python
%%sql 
SELECT b.NB, b.NOM, b.PRENOM FROM buveurs b
EXCEPT
SELECT DISTINCT b.NB, b.NOM, b.PRENOM FROM buveurs b, achats a, vins v
WHERE a.NB = b.NB and v.NV = a.NV and MILL = 1980
ORDER BY b.NB;
```

__2.9__ What clients (buveurs) have bought ONLY wines from 1980?

```python
%%sql
SELECT DISTINCT b.NB, b.NOM, PRENOM 
FROM buveurs as b, achats as a, vins as v
WHERE b.NB = a.NB AND a.NV = v.NV AND MILL = 1980
EXCEPT
SELECT DISTINCT b.NB, b.NOM, b.PRENOM
FROM buveurs as b, achats as a, vins as v
WHERE b.NB = a.NB AND a.NV = v.NV AND MILL <> 1980;
```
